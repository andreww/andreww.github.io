# Checkout

So, if you know the score, what are the possible checkouts? This works them all out for three darts ending on a double... 

First set up three variables for all the single dart scores, split into 'singles', 'doubles' and 'trebles'. Note that the bull is a special case.

```python
# All possible scores
singles = range(1, 21)
doubles = [s*2 for s in singles]
trebles = [s*3 for s in singles]

# Strings describing each score
singles_string = [str(s) for s in singles]
doubles_string = ['double ' + str(s) for s in singles]
trebles_string = ['treble ' + str(s) for s in singles]

# Special case for the bull...
singles.append(25) # Bullseye
doubles.append(50) # Double bullseye
singles_string.append('bullseye')
doubles_string.append('double bullseye')

# It is helpful to have a single list of all single dart scores (and descriptions) 
one_dart_scores = [0] + singles + doubles + trebles
one_dart_strings = ['-'] + singles_string + doubles_string + trebles_string
```

Now we can easily tabulate all possible 1, 2 and 3 dart outs and put them into a dictionary sorted by the score that is needed.

```python
outs = {}
for outscore in range(2,181):
    outs[outscore] = None
    for s_1, s_1s in zip(one_dart_scores, one_dart_strings): # All possible 1st dart scores
       for s_2, s_2s in zip(one_dart_scores, one_dart_strings): # All possible 2nd dart scores 
            # Final dart to go out
            for s_3, s_3s in zip(doubles, doubles_string):
                s = s_1 + s_2 + s_3
                if s == outscore:
                    if outs[s] is None:
                        outs[s] = [s_1s + ', ' + s_2s + ', ' + s_3s]
                    else:
                        outs[s].append(s_1s + ', ' + s_2s + ', ' + s_3s)
```

A function to return print the outs

```python
def dart_outs(score_needed):
    if ((score_needed <= 1 ) or (score_needed > 180) 
        or (outs[score_needed] is None)):
        print "No possible win in three darts"
    else:
        print "There are " + str(len(outs[score_needed])) + " ways to finish."
        print "There are:"
        for out in outs[score_needed]:
            print out
```

And an example of use

```python
dart_outs(143)
```

# Where to aim

Bit more involved. Given a measure of ability where should you aim? There are (clever) ways of doing this - see background material at http://www.stat.cmu.edu/~ryantibs/darts/ - but this shows how it can be done using a MC approach. First we need to import modules for fast maths and ploting.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

A function to give the score given an (x,y) position

```python
def score(x, y):
    """For a location (x, y) return dart's score
    
       x is the vertical distance above the center in mm,
       with arrows below the center of the bull having a 
       -ve value. y is the horizontal distance to the right 
       of the bull in mm, with arrows to the left of the bull
       having a -ve value.
       
       Size of board is taken from Appendix 1 of
       http://www.stat.cmu.edu/~ryantibs/papers/darts.pdf
    """
    
    # Convert x and y to theta and r as described above
    theta = np.arctan2(y, x) # in radians. note 'odd' order
    if theta < 0.0:
        theta = theta + 2.0*np.pi # So all angles are +ve clockwise
    r = np.hypot(x,y)
    
    # Given theta, work out the 'base' score
    # Start with theta in radians first convert
    # to fractions of a circle.
    theta_frac = theta/(2.0*np.pi) 
    if theta_frac < 0:
        raise ValueError
    elif theta_frac < 0.025:
        base_score = 20
    elif theta_frac < 0.075:
        base_score = 1
    elif theta_frac < 0.125:
        base_score = 18
    elif theta_frac < 0.175:
        base_score = 4
    elif theta_frac < 0.225:
        base_score = 13
    elif theta_frac < 0.275:
        base_score = 6
    elif theta_frac < 0.325:
        base_score = 10
    elif theta_frac < 0.375:
        base_score = 15
    elif theta_frac < 0.425:
        base_score = 2 
    elif theta_frac < 0.475:
        base_score = 17
    elif theta_frac < 0.525:
        base_score = 3
    elif theta_frac < 0.575:
        base_score = 19
    elif theta_frac < 0.625:
        base_score = 7
    elif theta_frac < 0.675:
        base_score = 16
    elif theta_frac < 0.725:
        base_score = 8
    elif theta_frac < 0.775:
        base_score = 11
    elif theta_frac < 0.825:
        base_score = 14
    elif theta_frac < 0.875:
        base_score = 9
    elif theta_frac < 0.925:
        base_score = 12
    elif theta_frac < 0.975:
        base_score = 5
    elif theta_frac < 1.0:
        base_score = 20
    else:
        raise ValueError
        
    # now compute the score from the radius
    if r < 6.35:
        score = 50 # double bull
    elif r < 15.9:
        score = 25 # outer bull
    elif r < 99.0:
        score = base_score
    elif r < 107.0:
        score = base_score * 3 # treble
    elif r < 162.0:
        score = base_score
    elif r < 170.0:
        score = base_score * 2 # double 
    else:
        score = 0 # missed the board
        
    return score
```

An example of use - aim at treble 20:

```python
score(100.0, 0.0)
```

Plot the results of randomly hitting the board

```python
# Check we get sensible looking results

# Put the darts randomly within a 170 mm of origin
N = 15000
r = 170.0 * np.sqrt(np.random.rand(N)) # sqrt to get equal area sampling
theta = 2.0*np.pi*np.random.rand(N)
# Convert to x and y 
x = r*np.cos(theta)
y = r*np.sin(theta)

# Calculate score
# ... should work out how to vectorize score, 
# but 'til then this crapness will do:
vec_score = np.vectorize(score)
colors = vec_score(x, y)

# For plotting
theta = np.arctan2(x, y)
r = np.hypot(x,y)
ax = plt.subplot(111, polar=True)
c = plt.scatter(theta, r, c=colors, s=10)
ax.set_rmax(170)
c.set_alpha(0.75)
plt.colorbar(c)

plt.show()
```

Which looks OK. Now we need a function to give a location given a aiming point.

```python
def calc_point(x_aim, y_aim, sigma, N):
    # Given an aiming point, (x,y) and 
    # and error term, sigma, find a 
    # point from a gaussian distribution
    x_hit = sigma * np.random.randn(N) + x_aim
    y_hit = sigma * np.random.randn(N) + y_aim
    return x_hit, y_hit
```

```python

# Aim at the treble 20 with N darts
N = 150
x_aim = 99.0+(0.5*(107.0-99.0))
y_aim = 0.0
sigma = 15.0
x, y = calc_point(x_aim, y_aim, sigma, N)

# Calculate score
# ... should work out how to vectorize score, 
# but 'til then this crapness will do:
vec_score = np.vectorize(score)
scores = vec_score(x, y)

# Report average score per dart
print "The mean score per dart is: " + str(np.sum(scores)/float(N))

# For plotting
theta = np.arctan2(x, y)
r = np.hypot(x,y)
ax = plt.subplot(111, polar=True)
c = plt.scatter(theta, r, c=scores, s=10)
ax.set_rmax(170)
c.set_alpha(0.75)
plt.colorbar(c)

plt.show()
```

Try somewhere close to treble 19

```python
# Aim at the treble 19 with N darts
N = 150
x_aim = -99.0-(0.5*(107.0-99.0))+5.0
y_aim = -22.0
sigma = 15.0
x, y = calc_point(x_aim, y_aim, sigma, N)

# Calculate score
# ... should work out how to vectorize score, 
# but 'til then this crapness will do:
vec_score = np.vectorize(score)
scores = vec_score(x, y)

# Report average score per dart
print "The mean score per dart is: " + str(np.sum(scores)/float(N))

# For plotting
theta = np.arctan2(x, y)
r = np.hypot(x,y)
ax = plt.subplot(111, polar=True)
c = plt.scatter(theta, r, c=scores, s=10)
ax.set_rmax(170)
c.set_alpha(0.75)
plt.colorbar(c)

plt.show()
```

So it looks like the crossover between aiming at the treble 20 and treble 19 is when sigma $\sim$ 15 mm. Lots of approximations!
